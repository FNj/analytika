## Shrnutí

Chceme publikovat objektivně nejlepší program, tj., proveditelný a rozumný program založený na kvalitních podkladech, datech a zpracovaný kvalifikovaným personálem. Aspirujeme na to, aby naše návrhy svou jednoduchostí a kvalitou zastínily návrhy ODS a ČSSD, vyplnili absenci konkrétního daňového programu u hnutí ANO, a ukázaly občanům třetí cestu: jde to i jednoduše.

Za prvé je třeba systém daní v České republice zjednodušit. Chceme systémově sjednotit zdanění práce (včetně odvodů na pojistné) do jedné sazby. Po daňové reformě Pirátů si bude moci každý spočítat svoje daně pomalu z hlavy. A pokud jde o čistou mzdu, bude výsledek po reformě jistě příznivější. Na potřebě zjednodušení se shodli i ekonomové [1].

Za druhé navrhujeme snížit celkové zdanění práce o 1,6 procentního bodu, aby zdanění práce v České republice již nepatřilo mezi nejvyšší mezi zeměmi OECD. Snížíme tak mzdové náklady firem, což přinese ekonomické oživení, zvýší zaměstnanost a zejména zvýší čisté mzdy 99 % zaměstnanců. Zaměstnanec s průměrným platem ušetří 7500 Kč ročně. Zdanění práce chceme dále postupně snižovat, ideálně o jeden procentní bod ročně. 


## Základní pojmy zdanění práce

Práce, stejně jako přírodní zdroje a kapitál, je jedním ze základních výrobních faktorů v ekonomice. Práce se obchoduje na trhu práce, kde zaměstnanci nabízejí svou práci a firmy (zaměstnavatelé) jejich práci poptávají za určitou cenu.

### Zdanění práce

**Zdaněním práce** rozumíme součet daně z příjmu a povinných odvodů na sociální a zdravotní pojištění, neboli všechny náklady na platby státu, které se promítají do ceny, za kterou zaměstnanec svou práci „pronajímá“ na trhu práce.

Z hlediska ekonomické rovnováhy na trhu práce je daň jen jedna nezávisle na tom, jak se její části deklarativně alokují mezi zaměstnance a firmu (současný zákon daň opticky snižuje tím, že ji rozděluje mezi firmu a zaměstnance, ve skutečnosti však o rozdělení břemena daně rozhodují tržní mechanismy vyjadřující elasticitu, nikoliv zákonný předpis). 

$$ \mathrm{zdanění\,práce\,} D = \mathrm{daň\,z\,příjmu\,} d + \mathrm{povinné\,pojistné\,} p $$

### Pojistné na povinné pojištění

**Pojistné** je specifické v tom, že zdravotní pojištěnec má právo na bezplatnou zdravotní péči a člověk platící sociální pojištění má právo na dávky sociálního pojištění jako je starobní důchod. V důchodu je zásluhovost zohledněna tím, že osoby platící vyšší sociální pojištění mají právo na vyšší důchod (ovšem se zvyšující se solidární složkou).

$$ \mathrm{povinné\,pojistné\,} p = \mathrm{zdravotní\,pojištění\,} p_\mathrm{z} + \mathrm{sociální\,pojištění\,} p_\mathrm{s} $$

Rozdělení zdravotního (1) a sociálního (2) pojištění mezi firmu a zaměstnance zachycujeme následovně:

$$
\begin{eqnarray*}
  p_1 &=& p_\mathrm{1f} + p_\mathrm{1z} \\
  p_2 &=& p_\mathrm{2f} + p_\mathrm{2z} 
\end{eqnarray*}
$$

Pojistné, které zákon opticky nechává platit zaměstnance je tedy 

$$ p_\mathrm{z} = p_\mathrm{1z} + p_\mathrm{2z} $$ 

a pojistné, které zákon nechává opticky platit firmu (zaměstnavatele) je rovno 

$$ p_\mathrm{f} = p_\mathrm{1f} + p_\mathrm{2f} $$ 

Příslušné sazby pojistného budeme značit veličinou $s$ s příslušným indexem, takže například sazba zdravotního pojištění na straně zaměstnance činí $s_\mathrm{1z}$ (podle dnes platného zákona). Výše pojistného se spočte ze mzdových nákladů $m$ (viz dále), takže při této sazbě bude výše pojistného $p_\mathrm{1z} = s_\mathrm{1z}m$.


### Čistá mzda

**Čistá mzda** $c$ je to, co zaměstnanec skutečně obdrží na svůj bankovní účet, tj. po stržení všech povinných plateb státu nebo veřejným zdravotním pojišťovnám. 

### Hrubá mzda

**Hrubá mzda** $h$ je to, co zaměstnanec vidí na výplatní pásce a co se dnes obvykle označuje pouze slovem mzda. Hrubá mzda dnes kvůli arbitrárnímu rozdělení sociálního a zdravotního pojištění mezi zaměstnance a firmu nemá přímý ekonomický význam. Pro pochopení konstrukce daňového systému je však důležitá, neboť se z ní vypočítává výše pojistného. 


### Mzdové náklady

**Mzdové náklady** $m$ jsou v podstatě cenou na trhu práce, která sestává jak z peněz jdoucí zaměstnanci (jeho čisté mzdy), tak zdanění jeho práce. Jde o klíčový ukazatel, podle kterého se firma rozhoduje, kolik práce najmout. Firmu totiž zajímá, kolik jí zaměstnanec bude stát (dnes superhrubá mzda). 

$$
\begin{eqnarray*}
\mathrm{mzdové\,náklady\,} m &=& \mathrm{čistá\,mzda\,} c + \mathrm{zdanění\,práce\,} D \\
&=& \mathrm{hrubá\,mzda\,} h  + \mathrm{pojistné\,na\,straně\,firmy\,} p_\mathrm{f} 
\end{eqnarray*}
$$

Druhá rovnost vyjadřuje, že mzdové náklady (to co firma musí zaplatit), tvoří pojistné na straně firmy (zaměstnavatele) a hrubá mzda zaměstnanci, přičemž ani jedna z těchto dvou posledně zmíněných veličin nemá přímý ekonomický význam.

Mezi hrubou mzdou a mzdovými náklady lze v současném systému přepočítávat pomocí koeficientu $m=kh$. Je tomu tak proto, že pojistné na straně zaměstnance vystupující v rovnici \ref{hruba-mzda} se odvíjí od hrubé mzdy. 
Konkrétně podle platné úpravy je $p_\mathrm{f} = 0,34h$ (zaměstnavatel platí 25 % z hrubé mzdy sociální pojištění a 9 % z hrubé mzdy zdravotní pojištění, tedy celkem 34 % z hrubé mzdy), a tedy $k=1,34$ pro hrubé mzdy menší než je čtyřnásobek průměrné mzdy. Z hrubé mzdy nad tuto částku už se neplatí sociální pojištění (jde o tzv. strop na pojistné na sociální pojištění), a tedy $p_\mathrm{f} = 0,09h$.



### Sazba zdanění práce

**Sazba zdanění práce** je podíl zdanění práce a mzdových nákladů. Vyjadřuje, kolik peněz zaplacených zaměstnavatelem musí být odvedeno státu:

$$ \mathrm{sazba\,zdanění\,práce\,} S = \frac{\mathrm{zdanění\,práce\,} D}{\mathrm{mzdové\,náklady\,} m} $$

Sazbu zdanění práce vyjadřujeme v procentech. Sazba zdanění práce se v tomto smyslu týká celého příjmu daného zaměstnance, jde vlastně o průměrnou sazbu zdanění práce při daných mzdových nákladech.

### Mezní sazba zdanění práce

**Mezní sazba zdanění práce** vyjadřuje, jakým procentem budu muset zdanit dodatečnou korunu, kterou mi zaměstnavatel zaplatí. Mezní sazba zdanění práce při dané výši příjmu je tedy poměrem mezi navýšením daně a navýšením příjmu:

$$ \mathrm{mezní\,sazba\,zdanění\,práce\,} s = \frac{\mathrm{navýšení\,zdanění\,práce\,} \Delta D}{\mathrm{navýšení\,mzdových\,nákladů\,} \Delta m} $$

Mezní sazba zdanění práce je klíčový pojem pro konstrukci daňových systémů. Mezní sazba může být stejná při libovolné výši příjmu, případně lidé s vyšším příjmem mohou platit vyšší mezní sazbu, potom jde o progresivní daň. Popis mezních sazeb plně popisuje daňový systém.

## Obecná konstrukce daňového systému

Daňový systém je pro naše účely jednoznačně zadán stanovením mezní sazby zdanění práce $s$ v závislosti na výši mzdových nákladů $m$. Důležitou veličinou je sleva na dani $P$, tj. částka, do které se neplatí daň z příjmu.

V této části budeme uvažovat zaměstnance pracujícího na plný úvazek, který uplatňuje pouze slevu na poplatníka ve výši $P$, nikoliv např. slevu na manželku nebo dítě. Tyto změny lze upravit změnou $P$. Rovněž nebereme v úvahu daňové bonusy, které u nízkopříjmových skupin s dětmi mohou vést k tomu, že poplatník obdrží od státu peníze místo placení daně z příjmu.

V tomto zjednodušeném schématu lze daňový systém reprezentovat posloupností daňových pásem pro mzdové náklady, mezi nimiž jsou hranice $H_0 = 0, H_1, H_2, \dots $ a příslušnými mezními daňovými sazbami $s_0, s_1, s_2, \dots $. Vhodným nastavením sazeb lze vytvořit progresivní nebo degresivní systém.
